```python
from matplotlib import pyplot as plt
```

```python
years = [1960, 1970, 1980, 1990, 2000, 2010, 2020, 2030]
gdp =   [508.33,865.98,1950,3630, 5490,16000,33430,70120]
```

```python
# Creates a line chart , where years on x axis and gdp on y axis
plt.plot(years,gdp, color='red', marker='o',linestyle='solid')
plt.title("Nepals GDP")
plt.ylabel("GDP in million $")
plt.xlabel("years")
```
