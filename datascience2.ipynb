```python
# Data Aggregation and GroupBy
# The split-apply-combine pattern
import pandas as pd
df = pd.read_csv("C:\\Users\\A plus\\Desktop\\1.B.tech\\5th Semester\\Data Science\\DS AI ML\\EmployeeID.csv")
df
```

```python
# Basic Grouping
# Group by department and calculate mean salary
df.groupby('Department')['Salary'].mean()
```

```python
# Multiple Aggregations:

df.groupby('Department').agg({
    'Salary': ['mean', 'min', 'max', 'std'],
    'Age' : 'mean'
})
```

```python
employee_stats = df.groupby('Department').agg({
    'Salary' : 'mean',
    'Age' : ['min', 'max'],
    'EmployeeID' : 'count'
})
print(employee_stats)
```

```python
# Merging and Combining Data
# Concatenation
# Example : split your main data in two smaller DataFrames
df1 = df.iloc[:5] # Start 0 to 5
df2 = df.iloc[5:] #start 5 to end

# Vertical concatenation (stack rows)
vertical_concat = pd.concat([df1, df2])

# Horizontal concatenation (stack columns)
# creating df3 as Dataframe of ratings

df3 = df[['Rating']]
horizontal_concat = pd.concat([df1.reset_index(drop=True), df3.reset_index(drop=True)])

print("\n Horizonatal Concatenation: \n", horizontal_concat)
```

```python
## Merging (SQL-like joins)
import pandas as pd
# Employee data
df = pd.DataFrame({
'EmployeeID': [101,102,103,104,105,106,107,108,109,110,111],
'Department': ['Sales','Marketing','Sales','Engineering','Marketing','Engineering','Sales','Engineerin','Marketing','Engineering','Sales'],
'Salary': [60000,55000,75000,90000,62000,110000,65000,95000,58000,10500,78888],
'Age': [35,28,42,30,31,45,25,33,29,38,28],
'Rating': ['A','B','A','C','A','B','B','A','C','A','B']
})
# Example: create a separate DataFrame with EmployeeID and some extra info 
bonus_df = pd.DataFrame({
'EmployeeID': [101,103,105,106,108,110],
'Bonus': [5000,7000,6000,10000,8000,9000]
})

# Inner join on EmployeeID
inner_merged = pd.merge(df, bonus_df, on='EmployeeID', how='inner')
print("Inner Join:\n", inner_merged)
# Left join (all employees, add bonus if exists)
left_merged = pd.merge(df, bonus_df, on='EmployeeID', how='left')
print("\nLeft Join:\n", left_merged)
# Full outer join (all employees + all bonus info)
outer_merged = pd.merge(df, bonus_df, on='EmployeeID', how='outer')
```
